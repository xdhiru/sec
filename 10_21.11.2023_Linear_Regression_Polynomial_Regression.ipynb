```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

head_data=pd.read_csv(f"{src}/headbrain.csv")
print(head_data)
```

```python
x=head_data["Head Size(cm^3)"].values
y=head_data["Brain Weight(grams)"].values
x_mean=np.mean(x)
y_mean=np.mean(y)

numerator=0
denominator=0
for i in range(len(x)):
    numerator += (x[i]-x_mean)*(y[i]-y_mean)
    denominator += (x[i]-x_mean)**2
m=numerator/denominator
c=y_mean-m*x_mean
print(m)
print(c)
```

```python
y1=m*x+c
print(y1)
```

```python
plt.plot(x, y1, color="r", label="regression line")
plt.scatter(x, y, color="b", label="Scatter Points")
plt.xlabel("Head Size")
plt.ylabel("Brain Size")
plt.show()
```

```python
num1 = 0
den1 = 0
for i in range(len(x)):
    y_pred = m*x[i] + c
    num1 +=(y[i] - y_mean)**2
    den1 +=(y[i] - y_pred)**2
r2 = 1-(den1/num1)
print(r2)

```

<h1>Automatic Methods (Inbuilt Tools)</h1>
<h2>  >  slope, intercept, r, p, std_err = stats.linregress(X, y)</h2>

```python
from scipy import stats
slope, intercept, r, p, std_err = stats.linregress(x, y)
print(intercept)
print(slope)
```

```python
def myfunc(xyz):
    return slope*xyz + intercept

y_pred = list(map(myfunc,x))

plt.scatter(x,y)
plt.plot(x, y_pred)
plt.show()
```

#Polynomial Regression

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]
pd.Series(y,index=x)
```

```python
y_pred=np.poly1d(np.polyfit(x,y,4))
y_pred
```

```python
plt.scatter(x, y)
plt.plot(x, y_pred(x))
plt.show()
```

```python
from sklearn.metrics import r2_score
print(r2_score(y,y_pred(x)))
```
