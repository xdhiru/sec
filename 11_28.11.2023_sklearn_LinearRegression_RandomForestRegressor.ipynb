```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
salary=pd.read_csv(f"{src}/Salary_Data.csv")
salary
```

```python
x=salary['YearsExperience']
x=pd.DataFrame(x)
x
```

```python
y=salary['Salary'].values
y
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
```

```python
model1=LinearRegression()
model2=RandomForestRegressor()
```

```python
model1.fit(x,y)
```

```python
model2.fit(x,y)
```

```python
y
```

```python
model1.predict(x)
```

```python
model2.predict(x)
```

```python
model1.score(x,y)
```

```python
model2.score(x,y)
```

```python
pd.concat([x,pd.DataFrame(y,columns=["Salary"]),pd.DataFrame(model1.predict(x),columns=["LinearRegression"]),pd.DataFrame(model2.predict(x),columns=["RandomForestRegressor"])],axis=1)
```

```python
plt.scatter(x,y)
plt.plot(x,model1.predict(x), color="g", label="LinearRegression (95.69%)")
plt.plot(x,model2.predict(x), color="r", label=f"RandomForestRegressor (99.19%)")
plt.legend()
plt.title("Accuracy comparision")
plt.show()
```
